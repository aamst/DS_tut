The session timestamp was saved to the JSON file as the number of miliseconds since the epoch. We need to convert that number back to a datetime in order to do proper analysis with it. 

```python
from datetime import datetime, date

import math

def deriveDateTimeFromEpochMS(ts):
    try:    
        if (isinstance(ts,float)):
            result = datetime.fromtimestamp(ts/1000.0)
            return result
        if (isinstance(ts,datetime)):
            return ts
    except:
        return datetime(2018, 10, 27, 12, 0)

ss['session_datetime']=ss['session_timestamp'].apply(deriveDateTimeFromEpochMS)
```
